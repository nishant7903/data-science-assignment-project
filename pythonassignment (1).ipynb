```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("C:\\Users\\nisha\\Downloads\\sample_data.csv")

# Display basic information and the first few rows
data.info()
data.head()

# Data Cleaning
# Convert date columns to datetime format (if any)
if 'date_column_name' in data.columns:  # Replace 'date_column_name' with the actual date column name
    data['date_column_name'] = pd.to_datetime(data['date_column_name'], errors='coerce')

# Drop non-numeric columns for correlation calculation
numeric_data = data.select_dtypes(include=[np.number])

# Drop rows with NaN values in numeric columns to avoid issues
numeric_data = numeric_data.dropna()

# Data Exploration
# Correlation Analysis
correlation_matrix = numeric_data.corr()
print(correlation_matrix)

# Plotting the Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Numeric Variables")
plt.show()

# Data Visualization Examples
# Scatter plot example
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=data)  # Adjust column names as needed
plt.title("Sepal Length vs Petal Length")
plt.show()

# Histogram example
data['sepal_length'].hist(bins=20)
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Length")
plt.show()

# Add titles to plot
plt.title("Your Plot Title Here")

# Axis labels
plt.xlabel("Your X-axis Label Here")
plt.ylabel("Your Y-axis Label Here")

# Save the notebook
!jupyter nbconvert --to notebook "final_assignment.ipynb"
```
